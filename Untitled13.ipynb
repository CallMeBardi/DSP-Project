```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# داده‌های برآوردی از اهمیت ویژگی (مقیاس‌گذاری شده در ۱۰۰۰)
importance_data = {
    'Band': ['Low']*4 + ['Mid']*4 + ['High']*4 + ['EHF']*4,
    'Bandwidth': ['8kHz', '12kHz', '16kHz', 'Full']*4,
    'Speech_Normal': [320, 310, 290, 280, 180, 200, 210, 220, 60, 80, 100, 130, 5, 15, 30, 50],
    'Speech_Strong': [340, 325, 300, 285, 200, 215, 225, 235, 70, 90, 115, 145, 10, 20, 40, 65],
    'Song_Normal': [310, 300, 280, 270, 160, 180, 195, 205, 50, 70, 90, 120, 8, 18, 35, 55],
    'Song_Strong': [330, 315, 290, 275, 185, 200, 215, 225, 65, 85, 110, 140, 12, 22, 45, 70]
}

df_imp = pd.DataFrame(importance_data)
df_imp = df_imp.melt(id_vars=['Band', 'Bandwidth'], 
                     var_name='Condition', value_name='Importance')
df_imp[['Modality', 'Intensity']] = df_imp['Condition'].str.split('_', expand=True)

# رسم
g = sns.FacetGrid(df_imp, col='Modality', row='Intensity', height=4, aspect=1.2)
g.map_dataframe(sns.barplot, x='Bandwidth', y='Importance', hue='Band', 
                palette='viridis', dodge=True, ci=None)
g.add_legend()
g.set_axis_labels('Bandwidth', 'Feature Importance (scaled x1000)')
g.fig.suptitle('Figure 2: Estimated Spectral Feature Importance by Band', y=1.02)
plt.tight_layout()
plt.show()
```

```python
# مقادیر Δ-distance برای شرایط مختلف (بر اساس متن)
delta_data = {
    'Condition': ['Speech_Normal', 'Speech_Strong', 'Song_Normal', 'Song_Strong'],
    '8kHz': [-0.008, -0.001, -0.001, 0.000],
    '12kHz': [-0.003, 0.000, 0.000, 0.001],
    '16kHz': [0.001, 0.001, 0.000, -0.001]
}

df_delta = pd.DataFrame(delta_data).melt(id_vars='Condition', 
                                         var_name='Bandwidth', 
                                         value_name='Delta_Distance')
df_delta[['Modality', 'Intensity']] = df_delta['Condition'].str.split('_', expand=True)

# رسم
plt.figure(figsize=(10, 6))
sns.barplot(data=df_delta, x='Bandwidth', y='Delta_Distance', 
            hue='Condition', palette='coolwarm', ci=None)
plt.axhline(0, color='black', linestyle='--', linewidth=0.8)
plt.ylabel('Δ Cosine Distance (relative to Full-band)')
plt.title('Figure 3: Overall Emotion-Space Separation Relative to Full-band')
plt.legend(title='Modality_Intensity')
plt.tight_layout()
plt.show()
```

```python
# داده‌های Δ برای جفت‌های احساسی در گفتار نرمال (قابل‌توجه‌ترین اثرات)
pairwise_data = {
    'Emotion Pair': ['Fearful-Happy', 'Angry-Happy', 'Happy-Sad', 'Angry-Sad', 'Angry-Fearful'],
    '8kHz': [-0.048, -0.027, -0.014, 0.009, 0.007],
    '12kHz': [-0.005, -0.003, -0.001, 0.002, 0.001],
    '16kHz': [0.001, 0.000, 0.001, 0.000, 0.000]
}

df_pair = pd.DataFrame(pairwise_data).melt(id_vars='Emotion Pair', 
                                           var_name='Bandwidth', 
                                           value_name='Delta_Cosine')

plt.figure(figsize=(12, 6))
sns.barplot(data=df_pair, x='Emotion Pair', y='Delta_Cosine', 
            hue='Bandwidth', palette='Set2', ci=None)
plt.axhline(0, color='black', linestyle='--', linewidth=0.8)
plt.ylabel('Δ Cosine Distance (relative to Full-band)')
plt.title('Figure 4: Changes in Pairwise Emotion Separability (Speech, Normal Intensity)')
plt.xticks(rotation=45)
plt.legend(title='Bandwidth')
plt.tight_layout()
plt.show()
```
